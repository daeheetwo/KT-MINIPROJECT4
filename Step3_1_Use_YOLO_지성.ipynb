#### 1) Label 파일의 클래스 값 변경

```python
base_folder = '/content/myface'  # 'my_face5' 폴더 경로
subfolders = ['train', 'valid', 'test']  # 하위 폴더 이름
unified_class = 1  # 통일할 클래스 번호

# 통일된 클래스로 변경 및 파일 수정
def unify_labels(base_folder, subfolders, unified_class=1):
    for subfolder in subfolders:
        labels_folder = os.path.join(base_folder, subfolder, 'labels')  # labels 폴더 경로
        txt_files = glob.glob(os.path.join(labels_folder, '*.txt'))  # 모든 .txt 파일 읽기

        for txt_file in txt_files:
            # 레이블 파일 읽기
            with open(txt_file, 'r') as file:
                lines = file.readlines()

            # 클래스 번호를 통일하여 새 내용 생성
            unified_lines = []
            for line in lines:
                parts = line.strip().split()
                parts[0] = str(unified_class)  # 클래스 번호 통일
                unified_lines.append(' '.join(parts))

            # 수정된 내용을 동일한 파일에 덮어쓰기
            with open(txt_file, 'w') as file:
                file.write('\n'.join(unified_lines))

# 하위 폴더의 레이블 통일
unify_labels(base_folder, subfolders, unified_class)
```

```python
import os
import glob

# 레이블 확인 함수
def check_unified_labels(base_folder, subfolders, expected_class=1):
    for subfolder in subfolders:
        labels_folder = os.path.join(base_folder, subfolder, 'labels')  # labels 폴더 경로
        txt_files = glob.glob(os.path.join(labels_folder, '*.txt'))  # 모든 .txt 파일 읽기

        for txt_file in txt_files:
            # 레이블 파일 읽기
            with open(txt_file, 'r') as file:
                lines = file.readlines()

            # 각 레이블 파일의 클래스 번호 확인
            for line in lines:
                parts = line.strip().split()
                if parts:  # 비어있지 않은 경우
                    class_id = int(parts[0])
                    if class_id != expected_class:
                        print(f"Unexpected class {class_id} found in file: {txt_file}")
                        break  # 발견 즉시 반복 종료
            else:
                print(f"All classes in file {txt_file} are as expected: {expected_class}")

# 통일된 레이블 확인
check_unified_labels(base_folder, subfolders, unified_class)
```

```python
base_folder = '/content/data_face2'  # datasets 폴더 경로
subfolders = ['train', 'valid', 'test']  # 하위 폴더 이름
unified_class = 0  # 통일할 클래스 번호

# 통일된 클래스로 변경 및 파일 수정
def unify_labels(base_folder, subfolders, unified_class=0):
    for subfolder in subfolders:
        labels_folder = os.path.join(base_folder, subfolder, 'labels')  # labels 폴더 경로
        txt_files = glob.glob(os.path.join(labels_folder, '*.txt'))  # 모든 .txt 파일 읽기

        for txt_file in txt_files:
            # 레이블 파일 읽기
            with open(txt_file, 'r') as file:
                lines = file.readlines()

            # 클래스 번호를 통일하여 새 내용 생성
            unified_lines = []
            for line in lines:
                parts = line.strip().split()
                parts[0] = str(unified_class)  # 클래스 번호 통일
                unified_lines.append(' '.join(parts))

            # 수정된 내용을 동일한 파일에 덮어쓰기
            with open(txt_file, 'w') as file:
                file.write('\n'.join(unified_lines))

# 하위 폴더의 레이블 통일
unify_labels(base_folder, subfolders, unified_class)
```

```python
import os
import shutil

# 각 폴더 경로 설정
#data_face1_base_dir = '/content/data_face1'
data_face2_base_dir = '/content/data_face2'
myface_base_dir = '/content/my_face5'
dataset_base_dir = '/content/datasets'  # 통합할 데이터셋 폴더

# 데이터셋 폴더가 없으면 생성
os.makedirs(os.path.join(dataset_base_dir, 'images', 'train'), exist_ok=True)
os.makedirs(os.path.join(dataset_base_dir, 'images', 'valid'), exist_ok=True)
os.makedirs(os.path.join(dataset_base_dir, 'images', 'test'), exist_ok=True)
os.makedirs(os.path.join(dataset_base_dir, 'labels', 'train'), exist_ok=True)
os.makedirs(os.path.join(dataset_base_dir, 'labels', 'valid'), exist_ok=True)
os.makedirs(os.path.join(dataset_base_dir, 'labels', 'test'), exist_ok=True)

# 데이터셋 통합 함수
def copy_dataset(source_dir, target_dir, subfolder):
    # 이미지 복사
    image_source = os.path.join(source_dir, subfolder, 'images')
    label_source = os.path.join(source_dir, subfolder, 'labels')

    if os.path.exists(image_source):
        shutil.copytree(image_source,
                        os.path.join(target_dir, 'images', subfolder),
                        dirs_exist_ok=True)
    if os.path.exists(label_source):
        shutil.copytree(label_source,
                        os.path.join(target_dir, 'labels', subfolder),
                        dirs_exist_ok=True)

# 각 데이터셋의 하위 폴더를 통합
for subfolder in ['train', 'valid', 'test']:
    #copy_dataset(data_face1_base_dir, dataset_base_dir, subfolder)
    copy_dataset(data_face2_base_dir, dataset_base_dir, subfolder)
    copy_dataset(myface_base_dir, dataset_base_dir, subfolder)

print("세 데이터셋이 성공적으로 통합되었습니다.")
```

```python
import yaml
import os

# 데이터셋 경로 설정
base_dir = '/content/datasets'
train_images_dir = os.path.join(base_dir, 'images', 'train')
val_images_dir = os.path.join(base_dir, 'images', 'valid')
test_images_dir = os.path.join(base_dir, 'images', 'test')

# YAML 파일에 저장할 데이터셋 구조 정의
dataset_config = {
    'path': base_dir,  # 데이터셋의 기본 경로
    'train': 'images/train',  # 훈련 이미지 상대 경로
    'val': 'images/valid',  # 검증 이미지 상대 경로
    'test': 'images/test',  # 테스트 이미지 상대 경로
    'nc': 2,  # 클래스 수
    'names': ['other_face', 'my_face']  # 클래스 이름 목록
}

# YAML 파일로 저장
output_yaml_path = os.path.join(base_dir, 'datasets.yaml')
with open(output_yaml_path, 'w') as yaml_file:
    yaml.dump(dataset_config, yaml_file, default_flow_style=False)

print("YAML 파일이 성공적으로 생성되었습니다:", output_yaml_path)
```
