```python
import os
import shutil
import random
from glob import glob
from zipfile import ZipFile

# ì••ì¶• íŒŒì¼ê³¼ í´ë” ê²½ë¡œ
data_myface = path + '/Datasets/YOLO/image.zip'
extract_folder = '/content/my_face'
output_folder = '/content/my_face'
split_ratio = (0.6, 0.2, 0.2)

# ì••ì¶• í•´ì œ
if not os.path.exists(extract_folder):
    os.makedirs(extract_folder)

with ZipFile(data_myface, 'r') as zip_ref:
    file_list = zip_ref.namelist()
    for f in file_list:
        if not f.endswith('/') and (f.lower().endswith('.jpg') or f.lower().endswith('.txt')):
            file_name = os.path.basename(f)
            if not file_name.startswith('._'):
                d_path = os.path.join(extract_folder, file_name)
                with zip_ref.open(f) as source, open(d_path, 'wb') as target:
                    target.write(source.read())

# íŒŒì¼ ëª©ë¡ ê°€ì ¸ì˜¤ê¸° ë° ìŒìœ¼ë¡œ ë¬¶ê¸°
images = sorted(glob(os.path.join(extract_folder, '*.jpg')))
labels = sorted(glob(os.path.join(extract_folder, '*.txt')))
data_pairs = list(zip(images, labels))

# ë°ì´í„° ì„ê¸°
random.shuffle(data_pairs)

# ë°ì´í„° ê°œìˆ˜ ê³„ì‚°
total_files = len(data_pairs)
train_count = int(total_files * split_ratio[0])
valid_count = int(total_files * split_ratio[1])
test_count = total_files - train_count - valid_count

# ë°ì´í„° ë¶„í• 
train_data = data_pairs[:train_count]
valid_data = data_pairs[train_count:train_count + valid_count]
test_data = data_pairs[train_count + valid_count:]

# í´ë” ìƒì„±
for split in ['train', 'valid', 'test']:
    os.makedirs(os.path.join(output_folder, split, 'images'), exist_ok=True)
    os.makedirs(os.path.join(output_folder, split, 'labels'), exist_ok=True)

# íŒŒì¼ ì´ë™ í•¨ìˆ˜
def move_files(data, split):
    for img_path, label_path in data:
        img_name = os.path.basename(img_path)
        label_name = os.path.basename(label_path)

        # ì´ë™ ê²½ë¡œ
        target_img_path = os.path.join(output_folder, split, 'images', img_name)
        target_label_path = os.path.join(output_folder, split, 'labels', label_name)

        # íŒŒì¼ ë³µì‚¬
        shutil.copy2(img_path, target_img_path)
        shutil.copy2(label_path, target_label_path)

# ë°ì´í„° ì´ë™
move_files(train_data, 'train')
move_files(valid_data, 'valid')
move_files(test_data, 'test')

print("ë°ì´í„°ê°€ ì„±ê³µì ìœ¼ë¡œ ë¶„í• ë˜ê³  í´ë” êµ¬ì¡°ì— ì €ì¥ë˜ì—ˆìŠµë‹ˆë‹¤.")
```

```python
import os
import shutil
from glob import glob

# ì›ë³¸ í´ë” ê²½ë¡œë“¤
#source_folders = ['/content/my_face_2', '/content/other_face_1', '/content/other_face_2']
source_folders = ['/content/my_face', '/content/other_face_2']

# ìƒˆë¡œìš´ í´ë” ê²½ë¡œ ì„¤ì •
total_face_folder = '/content/total_face4'

# í•˜ìœ„ í´ë” êµ¬ì¡° ìƒì„±
for main_folder in ['images', 'labels']:
    for split in ['train', 'val', 'test']:  # valid ëŒ€ì‹  valë¡œ ë³€ê²½
        os.makedirs(os.path.join(total_face_folder, main_folder, split), exist_ok=True)

# íŒŒì¼ ë³‘í•© í•¨ìˆ˜ ì •ì˜
def merge_files(source_folder, main_folder, source_split, target_split):
    target_folder = os.path.join(total_face_folder, main_folder, target_split)
    source_path = os.path.join(source_folder, source_split, main_folder)

    # ì§€ì •ëœ ê²½ë¡œì—ì„œ jpg/txt íŒŒì¼ì„ ë³µì‚¬
    file_extension = '*.jpg' if main_folder == 'images' else '*.txt'
    for file in glob(os.path.join(source_path, file_extension)):
        shutil.copy2(file, os.path.join(target_folder, os.path.basename(file)))

# ê° ì†ŒìŠ¤ í´ë”ì˜ train, valid, test í´ë”ë¥¼ total_face í´ë”ì— í†µí•©
for source in source_folders:
    for source_split, target_split in [('train', 'train'), ('valid', 'val'), ('test', 'test')]:
        merge_files(source, 'images', source_split, target_split)
        merge_files(source, 'labels', source_split, target_split)

print("ë°ì´í„°ê°€ total_face í´ë”ì— ì„±ê³µì ìœ¼ë¡œ ë³‘í•©ë˜ì—ˆìŠµë‹ˆë‹¤.")
```

```python
from ultralytics import YOLO

# ëª¨ë¸ í•™ìŠµ
model = YOLO("yolo11s.pt")  # load a pretrained model (recommended for training)

results = model.train(data="/content/total_face_dataset.yaml", epochs=1, imgsz=640, device=0)
```
